```python
data_by_zip = {}    
```

```python
new_attrs = ['zipcode', 'zipcode_type', 'major_city', 'post_office_city', 'county', 'state', 'lat', 'lng', 'timezone', 'radius_in_miles', 'population', 'population_density', 'land_area_in_sqmi', 'water_area_in_sqmi', 'housing_units', 'occupied_housing_units', 'median_home_value', 'median_household_income', 'bounds_west', 'bounds_east', 'bounds_north', 'bounds_south']
```

```python
search = SearchEngine(simple_zipcode=True, db_file_dir='data')
```

```python
print('----------start loop--------------')
for (i, zip) in enumerate(unique_zip):
    data = search.by_zipcode(zip).to_dict()
    data_by_zip.update({zip: data})
    if i % 1000:
        print('{} / {}'.format(i, len(unique_zip)))
    
```

```python
df_data_by_zip = pd.DataFrame(data_by_zip).transpose()
```
